```python
#6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
# uniform distribution.
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt

# Parameters for discrete uniform distribution (e.g., integers from 1 to 6)
a, b = 1, 6
values = np.arange(a, b + 1)

# Probability for each value (equal probabilities)
probabilities = np.ones_like(values) / len(values)

# Calculate the CDF as cumulative sum of PMF
cdf = np.cumsum(probabilities)

# Plot the CDF
plt.step(values, cdf, where='post')
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.ylim(0, 1.05)
plt.grid(True)
plt.show()
```

```python
#15. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 5          # population mean
sigma = 2       # population standard deviation
sample_size = 30  # size of each sample
num_samples = 1000  # number of samples

# Generate samples and compute their means
sample_means = []
for _ in range(num_samples):
    sample = np.random.normal(mu, sigma, sample_size)
    sample_means.append(np.mean(sample))
sample_means = np.array(sample_means)

# Calculate mean and std deviation of sample means
mean_of_sample_means = np.mean(sample_means)
std_of_sample_means = np.std(sample_means)

# Plot histogram of sample means
plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')

# Overlay normal distribution curve with mean and std of sample means
x = np.linspace(min(sample_means), max(sample_means), 100)
plt.plot(x, norm.pdf(x, mean_of_sample_means, std_of_sample_means), 'r', linewidth=2)

plt.title('CLT Verification: Sample Means from Normal Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.grid(True)
plt.show()

print(f"Mean of sample means: {mean_of_sample_means:.3f}")
print(f"Std dev of sample means: {std_of_sample_means:.3f}")
print(f"Theoretical std dev: {sigma / np.sqrt(sample_size):.3f}")
```

```python
#19.  Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
from scipy.stats import norm


def z_test(sample, population_mean, population_std, alpha=0.05):
    n = len(sample)
    sample_mean = np.mean(sample)

    # Calculate standard error
    std_error = population_std / np.sqrt(n)

    # Calculate Z-statistic
    z_stat = (sample_mean - population_mean) / std_error

    # Calculate p-value for two-tailed test
    p_value = 2 * (1 - norm.cdf(abs(z_stat)))

    # Determine if we reject the null hypothesis
    reject_null = p_value < alpha

    print(f"Sample mean = {sample_mean:.3f}")
    print(f"Z-statistic = {z_stat:.3f}")
    print(f"P-value = {p_value:.4f}")
    print(f"Reject null hypothesis? {'Yes' if reject_null else 'No'}")

    return z_stat, p_value, reject_null

# Example data and known population parameters
sample_data = np.array([102, 100, 98, 101, 99, 97, 103, 100, 96, 102])
population_mean = 100
population_std = 5  # known population std deviation

# Run the Z-test
z_test(sample_data, population_mean, population_std)
```

```python
#20. Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # Standard error of the mean
    t_critical = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t-critical value

    margin_of_error = t_critical * std_err
    ci_lower = mean - margin_of_error
    ci_upper = mean + margin_of_error

    print(f"Sample mean = {mean:.3f}")
    print(f"{int(confidence*100)}% Confidence interval = [{ci_lower:.3f}, {ci_upper:.3f}]")

    return ci_lower, ci_upper

# Example dataset
data = np.array([12, 15, 14, 10, 8, 11, 13, 14, 16, 9])

# Calculate 95% confidence interval
confidence_interval(data)
```

```python
#27. Implement Z-score normalization on a dataset and visualize its transformation
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(loc=50, scale=10, size=500)

def z_score_normalize(dataset):
    mean = np.mean(dataset)
    std = np.std(dataset, ddof=1)
    return (dataset - mean) / std


normalized_data = z_score_normalize(data)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title('Original Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, color='salmon', edgecolor='black')
plt.title('Z-score Normalized Data')
plt.xlabel('Z-score')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
# distribution.
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(mu=0, sigma=1, sample_size=1000):

    data = np.random.normal(mu, sigma, sample_size)

    data_skewness = skew(data)
    data_kurtosis = kurtosis(data)

    print(f"Skewness: {data_skewness:.4f}")
    print(f"Kurtosis (excess): {data_kurtosis:.4f}")

    return data_skewness, data_kurtosis

calculate_skewness_kurtosis()
```
